## ⚙️ Imports & Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['figure.dpi'] = 100
sns.set_theme(style='whitegrid')

print('✅ Libraries loaded')
```

---
## 📂 1. Load PLC Data

```python
# *** UPDATE THIS PATH TO YOUR FILE ***
PLC_PATH = '../data/raw/plc_data.csv'

try:
    plc_df = pd.read_csv(PLC_PATH)
    print(f'✅ Loaded: {plc_df.shape[0]:,} rows × {plc_df.shape[1]} columns')
except FileNotFoundError:
    print(f'❌ File not found at: {PLC_PATH}')
    print('   Please update PLC_PATH above.')
    plc_df = None
```

---
## 📊 2. Data Overview
First look at columns, data types, and readings per coil.

```python
if plc_df is not None:
    print(f'Shape : {plc_df.shape}')
    print(f'Rows  : {plc_df.shape[0]:,}')
    print(f'Cols  : {plc_df.shape[1]}')
    print('\n📋 Column names, types and sample values:')
    dtype_df = pd.DataFrame({
        'Column'  : plc_df.columns,
        'Dtype'   : plc_df.dtypes.values,
        'Non-Null': plc_df.notnull().sum().values,
        'Sample'  : [plc_df[c].dropna().iloc[0] if plc_df[c].notnull().any() else 'N/A'
                     for c in plc_df.columns]
    })
    display(dtype_df)
```

```python
if plc_df is not None:
    coil_candidates = [c for c in plc_df.columns if any(k in c.upper() for k in ['COIL','BOBINA','ID','NUM'])]
    pos_candidates  = [c for c in plc_df.columns if any(k in c.upper() for k in ['MT','METER','POS','LEN','FROM','TO','DIST'])]
    date_candidates = [c for c in plc_df.columns if any(k in c.upper() for k in ['DATE','TIME','FECHA','DATA'])]
    print('🔎 Likely coil ID columns :', coil_candidates)
    print('🔎 Likely position columns:', pos_candidates)
    print('🔎 Likely date columns    :', date_candidates)
    print('\n⚠️  Review above, then set variables in the next cell.')
```

> ⚠️ **Action required:** After running the cell above, update the three variables below to match your actual column names.

```python
# *** SET THESE BASED ON THE OUTPUT ABOVE ***
COIL_COL = coil_candidates[0] if coil_candidates else 'COIL'
POS_COL  = pos_candidates[0]  if pos_candidates  else 'MT'
DATE_COL = date_candidates[0] if date_candidates else None

print(f"COIL_COL = '{COIL_COL}'")
print(f"POS_COL  = '{POS_COL}'")
print(f"DATE_COL = '{DATE_COL}'")
```

```python
if plc_df is not None and COIL_COL in plc_df.columns:
    n_coils       = plc_df[COIL_COL].nunique()
    rows_per_coil = plc_df.groupby(COIL_COL).size()
    print(f'Unique coils             : {n_coils:,}')
    print(f'Rows per coil — mean     : {rows_per_coil.mean():.1f}')
    print(f'Rows per coil — min/max  : {rows_per_coil.min()} / {rows_per_coil.max()}')
    print(f'Coils with only 1 row    : {(rows_per_coil == 1).sum()}')
    if DATE_COL and DATE_COL in plc_df.columns:
        plc_df[DATE_COL] = pd.to_datetime(plc_df[DATE_COL], errors='coerce')
        print(f'\nDate range : {plc_df[DATE_COL].min()} → {plc_df[DATE_COL].max()}')
    fig, ax = plt.subplots(figsize=(10, 4))
    rows_per_coil.clip(upper=rows_per_coil.quantile(0.99)).hist(bins=50, ax=ax, color='steelblue', edgecolor='white')
    ax.set_title('Distribution of PLC Readings per Coil')
    ax.set_xlabel('Number of rows')
    ax.set_ylabel('Number of coils')
    plt.tight_layout()
    plt.show()
```

---
## 🕳️ 3. Missing Data Analysis
Which sensors have missing values, and do the same rows tend to be missing together?

```python
if plc_df is not None:
    missing         = plc_df.isnull().sum()
    missing_pct     = (missing / len(plc_df)) * 100
    cols_missing    = missing[missing > 0].sort_values(ascending=False)
    missing_pct_flt = missing_pct[missing > 0].sort_values(ascending=False)
    print(f'Columns with NO missing data : {(missing == 0).sum()}')
    print(f'Columns WITH missing data    : {(missing > 0).sum()}')
    print(f'Overall missing rate         : {missing.sum() / plc_df.size * 100:.2f}%')
    if len(cols_missing) > 0:
        display(pd.DataFrame({'Missing Count': cols_missing, 'Missing %': missing_pct_flt.round(2)}).head(30))
        fig, ax = plt.subplots(figsize=(14, 5))
        missing_pct_flt.head(30).plot(kind='bar', ax=ax, color='salmon', edgecolor='white')
        ax.set_title('Top 30 Columns by Missing Data %')
        ax.set_ylabel('Missing %')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print('✅ No missing data found!')
```

```python
if plc_df is not None and 'cols_missing' in dir() and len(cols_missing) > 0:
    missing_cols_list = cols_missing.index.tolist()
    missing_matrix    = plc_df[missing_cols_list].isnull().astype(int)
    if len(missing_cols_list) > 1:
        fig, ax = plt.subplots(figsize=(12, 8))
        sns.heatmap(missing_matrix.corr(), cmap='RdYlGn_r', center=0, vmin=-1, vmax=1,
                    ax=ax, annot=len(missing_cols_list) <= 20, fmt='.1f', linewidths=0.5)
        ax.set_title('Missingness Correlation\n(1.0 = always missing together)')
        plt.tight_layout()
        plt.show()
```

---
## 📈 4. Sensor Distributions
Summary statistics and histograms for every numeric sensor.

```python
if plc_df is not None:
    exclude_cols = [c for c in [COIL_COL, POS_COL, DATE_COL] if c]
    sensor_cols  = [c for c in plc_df.select_dtypes(include=np.number).columns
                    if c not in exclude_cols]
    print(f'Numeric sensor columns: {len(sensor_cols)}')
    print(sensor_cols[:20], '...' if len(sensor_cols) > 20 else '')
    stats = plc_df[sensor_cols].describe().T
    stats['cv']       = (stats['std'] / stats['mean'].abs()).round(3)
    stats['zero_pct'] = ((plc_df[sensor_cols] == 0).sum() / len(plc_df) * 100).round(2)
    display(stats.sort_values('std', ascending=False).head(30))
```

```python
if plc_df is not None and 'sensor_cols' in dir() and len(sensor_cols) > 0:
    plot_cols = sensor_cols[:24]
    ncols = 4
    nrows = (len(plot_cols) + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows, ncols, figsize=(16, nrows * 3))
    axes = axes.flatten()
    for i, col in enumerate(plot_cols):
        data = plc_df[col].dropna()
        axes[i].hist(data.clip(data.quantile(0.01), data.quantile(0.99)), bins=50, color='steelblue', edgecolor='white')
        axes[i].set_title(col, fontsize=9)
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)
    fig.suptitle('Sensor Distributions (clipped 1st–99th percentile)', fontsize=13, y=1.01)
    plt.tight_layout()
    plt.show()
```

---
## 🔒 5. Stuck / Constant Sensors
Sensors with very few unique values are likely broken or irrelevant.

```python
if plc_df is not None and 'sensor_cols' in dir():
    unique_counts = plc_df[sensor_cols].nunique()
    stuck   = unique_counts[unique_counts <= 1]
    low_var = unique_counts[(unique_counts > 1) & (unique_counts <= 5)]
    print(f'Completely constant sensors : {len(stuck)}')
    if len(stuck): print(list(stuck.index))
    print(f'\nAlmost constant (≤5 values) : {len(low_var)}')
    if len(low_var):
        display(pd.DataFrame({
            'Unique values': low_var,
            'Value counts' : [str(plc_df[c].value_counts().head(3).to_dict()) for c in low_var.index]
        }))
    if len(stuck) == 0 and len(low_var) == 0:
        print('✅ No stuck or constant sensors found!')
```

---
## 📏 6. Sensor Behaviour Along Coil Length
How do sensor values evolve from start to end of a coil?

```python
if plc_df is not None and COIL_COL in plc_df.columns and POS_COL in plc_df.columns and 'sensor_cols' in dir():
    sample_coils = plc_df[COIL_COL].value_counts().head(5).index.tolist()
    plot_sensors = sensor_cols[:4]  # change to sensors of interest
    for coil in sample_coils[:3]:
        coil_data = plc_df[plc_df[COIL_COL] == coil].sort_values(POS_COL)
        fig, axes = plt.subplots(1, len(plot_sensors), figsize=(16, 3))
        if len(plot_sensors) == 1: axes = [axes]
        for ax, col in zip(axes, plot_sensors):
            ax.plot(coil_data[POS_COL], coil_data[col], linewidth=0.8, alpha=0.8)
            ax.set_title(col, fontsize=9)
            ax.set_xlabel('Position (m)')
        fig.suptitle(f'Coil {coil} — Sensor Profiles Along Length', fontsize=11)
        plt.tight_layout()
        plt.show()
```

---
## 🔗 7. Sensor Correlations
Identify redundant sensors — if two sensors have |r| ≥ 0.90, we may only need one.

```python
if plc_df is not None and 'sensor_cols' in dir() and len(sensor_cols) > 1:
    sample = plc_df[sensor_cols].sample(min(10_000, len(plc_df)), random_state=42)
    corr   = sample.corr()
    mask   = np.triu(np.ones_like(corr, dtype=bool))
    fig, ax = plt.subplots(figsize=(14, 11))
    sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0, vmin=-1, vmax=1,
                ax=ax, linewidths=0.3, annot=len(sensor_cols) <= 15, fmt='.2f')
    ax.set_title('Sensor Correlation Matrix')
    plt.tight_layout()
    plt.show()
    high_corr = []
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i, j]) >= 0.90:
                high_corr.append({'Sensor A': corr.columns[i], 'Sensor B': corr.columns[j], 'r': round(corr.iloc[i, j], 3)})
    if high_corr:
        print(f'⚠️  Highly correlated pairs (|r| ≥ 0.90): {len(high_corr)}')
        display(pd.DataFrame(high_corr).sort_values('r', key=abs, ascending=False))
    else:
        print('✅ No sensor pairs with |r| ≥ 0.90')
```

---
## ⚖️ 8. Process Stability — Within-Coil Variance
How much does each coil vary internally? Outlier coils may indicate production anomalies.

```python
if plc_df is not None and COIL_COL in plc_df.columns and 'sensor_cols' in dir():
    coil_std  = plc_df.groupby(COIL_COL)[sensor_cols].std()
    coil_mean = plc_df.groupby(COIL_COL)[sensor_cols].mean().abs()
    coil_cv   = coil_std.div(coil_mean)
    mean_cv_per_coil = coil_cv.mean(axis=1)
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    mean_cv_per_coil.hist(bins=50, ax=axes[0], color='mediumseagreen', edgecolor='white')
    axes[0].set_title('Mean Within-Coil CV')
    axes[0].set_xlabel('CV (std / |mean|)')
    mean_cv_per_coil.nlargest(10).plot(kind='barh', ax=axes[1], color='tomato')
    axes[1].set_title('Top 10 Most Variable Coils')
    axes[1].set_xlabel('Mean CV')
    plt.tight_layout()
    plt.show()
    print(f'Median within-coil CV : {mean_cv_per_coil.median():.3f}')
    print(f'Max within-coil CV    : {mean_cv_per_coil.max():.3f}  (coil: {mean_cv_per_coil.idxmax()})')
```

---
## 🔗 9. Merge Readiness Check
Verify the PLC data is ready to join with the defect labels.

```python
if plc_df is not None:
    checks = {}
    checks['COIL_COL exists']     = COIL_COL in plc_df.columns
    checks['POS_COL exists']      = POS_COL  in plc_df.columns
    if COIL_COL in plc_df.columns:
        checks['No null coil IDs']    = plc_df[COIL_COL].isnull().sum() == 0
    if POS_COL in plc_df.columns:
        checks['No null positions']   = plc_df[POS_COL].isnull().sum() == 0
        checks['Position is numeric'] = pd.api.types.is_numeric_dtype(plc_df[POS_COL])
    checks['No duplicate rows']   = plc_df.duplicated().sum() == 0
    print('PLC data checks:')
    for check, passed in checks.items():
        print(f"  {'✅' if passed else '❌'}  {check}")
    try:
        defect_df = pd.read_csv('../data/cleaned/defect_cleaned.csv')
        plc_coils = set(plc_df[COIL_COL].unique())
        def_coils = set(defect_df['COIL'].unique())
        overlap   = plc_coils & def_coils
        print(f'\n📊 Coil Overlap:')
        print(f'  PLC coils       : {len(plc_coils):,}')
        print(f'  Defect coils    : {len(def_coils):,}')
        print(f'  ✅ In both       : {len(overlap):,}')
        print(f'  ⚠️  PLC only     : {len(plc_coils - def_coils):,}')
        print(f'  ⚠️  Defect only  : {len(def_coils - plc_coils):,}')
    except FileNotFoundError:
        print('\n⚠️  defect_cleaned.csv not found — skipping overlap check.')
```
