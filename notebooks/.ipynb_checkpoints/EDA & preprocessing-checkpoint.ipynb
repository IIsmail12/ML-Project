```python
# Cell 14 improved - with readable labels
defect_cols = ['DIF_TIPO_1', 'DIF_TIPO_2', 'DIF_TIPO_3', 
               'DIF_TIPO_4', 'DIF_TIPO_5', 'DIF_TIPO_6']

# Create readable labels
def get_defect_names(row):
    active_defects = []
    for i, col in enumerate(defect_cols):
        if row[col] == 1:
            active_defects.append(f"Tipo{i+1}")
    return ', '.join(active_defects) if active_defects else 'No Defect'

defect_df['defects_present'] = defect_df.apply(get_defect_names, axis=1)

# Also keep binary for analysis
defect_matrix = defect_df[defect_cols].copy()
defect_matrix['combination_binary'] = defect_matrix.apply(
    lambda row: ''.join([str(int(x)) for x in row]), axis=1
)
defect_matrix['combination_readable'] = defect_df['defects_present']

print("📊 Defect combinations (top 10):")
print("-" * 60)
combo_counts = defect_matrix['combination_readable'].value_counts().head(10)
for combo, count in combo_counts.items():
    pct = (count / len(defect_df)) * 100
    print(f"{combo:30} : {count:4d} occurrences ({pct:5.2f}%)")
```

```python
# Cell 16: Check meter ranges
defect_df['segment_length'] = defect_df['MT_TO'] - defect_df['MT_FROM']

print("📏 Segment length statistics:")
print(f"  Min: {defect_df['segment_length'].min():.1f} meters")
print(f"  Max: {defect_df['segment_length'].max():.1f} meters")
print(f"  Mean: {defect_df['segment_length'].mean():.1f} meters")
print(f"  Median: {defect_df['segment_length'].median():.1f} meters")
```

```python
# Cell 15 alternative - Co-occurrence heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate co-occurrence matrix
cooccurrence = np.zeros((6, 6))
for i, col1 in enumerate(defect_cols):
    for j, col2 in enumerate(defect_cols):
        both = ((defect_df[col1] == 1) & (defect_df[col2] == 1)).sum()
        cooccurrence[i, j] = both

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cooccurrence, 
            annot=True, 
            fmt='d',
            xticklabels=['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4', 'Tipo 5', 'Tipo 6'],
            yticklabels=['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4', 'Tipo 5', 'Tipo 6'],
            cmap='YlOrRd')
plt.title('Defect Type Co-occurrence Matrix', fontsize=14, fontweight='bold')
plt.xlabel('Defect Type')
plt.ylabel('Defect Type')
plt.tight_layout()
plt.show()

print("💡 Darker colors = defects often occur together")
```

```python
# Calculate % of co-occurrence
total_defects = np.diag(cooccurrence).sum()
co_pairs = cooccurrence.sum() - np.diag(cooccurrence).sum()
co_pct = (co_pairs / (total_defects * 5)) * 100  # approximate

print(f"🔍 Key Finding:")
print(f"Most defects occur ALONE ({(100-co_pct):.1f}% of cases)")
print(f"Only {co_pct:.1f}% are combinations")
```

```python
# Cell: Which defect pairs are most common?
from itertools import combinations

# Find most common defect pairs
defect_pairs = []
for i, j in combinations(range(6), 2):
    pair_count = ((defect_df[defect_cols[i]] == 1) & 
                  (defect_df[defect_cols[j]] == 1)).sum()
    if pair_count > 0:
        defect_pairs.append({
            'pair': f'Tipo {i+1}+{j+1}',
            'count': pair_count
        })

pairs_df = pd.DataFrame(defect_pairs).sort_values('count', ascending=False)

plt.figure(figsize=(10, 6))
top_pairs = pairs_df.head(8)
bars = plt.barh(top_pairs['pair'], top_pairs['count'], color='coral', edgecolor='black')
plt.xlabel('Number of Occurrences', fontsize=12)
plt.title('Most Common Defect Pairs', fontsize=14, fontweight='bold')
for bar, val in zip(bars, top_pairs['count']):
    plt.text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2, 
             str(val), va='center', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n🔍 Top 3 defect pairs:")
for i, row in pairs_df.head(3).iterrows():
    print(f"  {row['pair']}: {row['count']} occurrences")
```

```python
# Cell 17b: Interpreting Segment Lengths
display(Markdown('''
### 📏 Understanding Defect Segment Lengths

The plots above show us:
- **How long are typical defect zones?**
- **Are there unusually long or short defects?**
- **Does this affect how we merge with PLC data?**
'''))

# Calculate key statistics
mean_length = defect_df['segment_length'].mean()
median_length = defect_df['segment_length'].median()
min_length = defect_df['segment_length'].min()
max_length = defect_df['segment_length'].max()
q1 = defect_df['segment_length'].quantile(0.25)
q3 = defect_df['segment_length'].quantile(0.75)
iqr = q3 - q1

print("📐 Segment Length Statistics:")
print(f"  Mean:     {mean_length:.1f} meters")
print(f"  Median:   {median_length:.1f} meters")
print(f"  Min:      {min_length:.1f} meters")
print(f"  Max:      {max_length:.1f} meters")
print(f"  Q1 (25%): {q1:.1f} meters")
print(f"  Q3 (75%): {q3:.1f} meters")
print(f"  IQR:      {iqr:.1f} meters")

# Identify potential outliers
outliers = defect_df[defect_df['segment_length'] > q3 + 1.5 * iqr]
print(f"\n🔍 Potential outliers: {len(outliers)} segments ({len(outliers)/len(defect_df)*100:.1f}%)")
if len(outliers) > 0:
    print(f"  Longest outlier: {outliers['segment_length'].max():.1f} meters")
```

```python
# Cell: Deep Dive into Defect Lengths
import numpy as np

print("🔍 DEFECT LENGTH DISTRIBUTION - DETAILED")
print("=" * 60)

# Basic stats
print("\n📊 Summary Statistics:")
print(defect_df['segment_length'].describe())

# Percentiles (to see distribution)
print("\n📈 Percentiles:")
percentiles = [1, 5, 10, 25, 50, 75, 90, 95, 99]
for p in percentiles:
    val = defect_df['segment_length'].quantile(p/100)
    print(f"  {p}th percentile: {val:.1f} meters")

# Check if "long" is actually long
print("\n🔍 What does 'long' mean?")
print(f"  Median length: {defect_df['segment_length'].median():.1f} meters")
print(f"  Mean length: {defect_df['segment_length'].mean():.1f} meters")

# Look at the distribution in more detail
print("\n📊 Length buckets:")
bins = [0, 10, 20, 30, 40, 50, 75, 100, 200, 500, 1000, 5000]
defect_df['length_bucket'] = pd.cut(defect_df['segment_length'], bins)
bucket_counts = defect_df['length_bucket'].value_counts().sort_index()
for bucket, count in bucket_counts.items():
    pct = count / len(defect_df) * 100
    print(f"  {bucket}: {count:4d} defects ({pct:5.1f}%)")
```

```python
# Cell: Possible explanations for "long" defects

# 1. Are there duplicate rows for the same defect?
print("🔍 Checking for potential issues:")
print("-" * 40)

# Check if same coil has overlapping defect ranges
coil_groups = defect_df.groupby('COIL')

overlap_count = 0
for coil, group in coil_groups:
    if len(group) > 1:
        # Sort by start meter
        group = group.sort_values('MT_FROM')
        for i in range(len(group)-1):
            if group.iloc[i]['MT_TO'] > group.iloc[i+1]['MT_FROM']:
                overlap_count += 1
                print(f"  Coil {coil}: Segment {group.iloc[i]['MT_FROM']}-{group.iloc[i]['MT_TO']} overlaps with {group.iloc[i+1]['MT_FROM']}-{group.iloc[i+1]['MT_TO']}")

if overlap_count == 0:
    print("  ✅ No overlapping defect segments found")

# 2. Check if some coils have unusually long defects
print("\n📏 Top 10 longest defect segments:")
longest = defect_df.nlargest(10, 'segment_length')[['COIL', 'MT_FROM', 'MT_TO', 'segment_length'] + defect_cols]
display(longest)

# 3. Check if some coils have many short defects
print("\n📏 Bottom 10 shortest defect segments (non-zero):")
shortest = defect_df[defect_df['segment_length'] > 0].nsmallest(10, 'segment_length')[['COIL', 'MT_FROM', 'MT_TO', 'segment_length'] + defect_cols]
display(shortest)
```

```python
# Cell: Do different defect types have different lengths?
print("📊 Average length by defect type:")
print("-" * 40)

for col in defect_cols:
    mask = defect_df[col] == 1
    avg_len = defect_df.loc[mask, 'segment_length'].mean()
    count = mask.sum()
    print(f"{col}: {avg_len:.1f} meters avg ({count} occurrences)")

# Visualize
plt.figure(figsize=(12, 6))
data_to_plot = [defect_df[defect_df[col] == 1]['segment_length'] for col in defect_cols]
plt.boxplot(data_to_plot, labels=defect_cols)
plt.title('Defect Length Distribution by Type', fontsize=14, fontweight='bold')
plt.xlabel('Defect Type')
plt.ylabel('Length (meters)')
plt.ylim(0, min(500, defect_df['segment_length'].quantile(0.95)))  # Cap to see detail
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Cell: Check if "defect" is actually the whole coil
print("🔍 Checking for coils marked as defective for entire length:")
print("-" * 60)

# Calculate total coil length (max meter per coil)
coil_max_meter = defect_df.groupby('COIL')['MT_TO'].max().reset_index()
coil_max_meter.columns = ['COIL', 'coil_total_length']

# Merge back
defect_with_coil_len = defect_df.merge(coil_max_meter, on='COIL')

# Find segments that cover >80% of the coil
defect_with_coil_len['pct_of_coil'] = (defect_with_coil_len['segment_length'] / defect_with_coil_len['coil_total_length']) * 100
whole_coil_defects = defect_with_coil_len[defect_with_coil_len['pct_of_coil'] > 80]

print(f"Found {len(whole_coil_defects)} segments that cover >80% of the coil!")
print("\nExamples:")
display(whole_coil_defects[['COIL', 'MT_FROM', 'MT_TO', 'segment_length', 'coil_total_length', 'pct_of_coil']].head(10))
```
